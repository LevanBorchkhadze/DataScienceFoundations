### Exercise 2 -  Calculating Correlation in Pandas


We will be using a dataset "Boston house prices". Calculate the Pearson correlation between "MEDV-Median value of owner-occupied homes in $1000's)" and all the variables.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets 
data = datasets.load_boston() 
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df.head()
```

```python
target = pd.DataFrame(data.target, columns=["MEDV"]) 
target.head()
```

```python
print (data.DESCR)
```

```python
df2 = df.assign(target=target.values)
df2.head()
```

```python
df2.rename(columns={'target': 'MEDV'}, inplace=True)
df2.head()
```

```python
df2.corr().style.format('{:.2}').background_gradient(cmap=plt.get_cmap('coolwarm'))
```

```python
corrs = pd.DataFrame(df2.corr()['MEDV'][0:13])
corrs.style.format('{:.2}').background_gradient(cmap=plt.get_cmap('coolwarm'))
```

```python
corrs.sort_values(by=['MEDV']).style.format('{:.2}').background_gradient(cmap=plt.get_cmap('coolwarm'))
```
