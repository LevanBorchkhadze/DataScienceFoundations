```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder  
%matplotlib inline
df=pd.read_csv("1_titanic_dataset.csv")
df.head()
```

```python
df.drop(['PassengerId'], axis=1).corr().style.format('{:.2}').background_gradient(cmap=plt.get_cmap('PuBu'), axis=1)
```

```python
corrs = pd.DataFrame(df.corr()['Survived'][2:])
corrs.sort_values(by='Survived').style.format('{:.2}').background_gradient(cmap=plt.get_cmap('PuBu'))
```

Most importatn variable in Titanic dataset is "Survived", it shows wherer passenger survived or not. Observing on the variable correlations with other items, we can say that only two of them ar correlated (Pcalss and Fare). Pclass has negative correlation, wchich according to data means that person in lower class had less probability of survival. Contrary Fare has possitive correlation, higher price person payed, higher was survival probability.

As we can see from results above, pandas corr() function couldn't consider gender and Embarked variables because they are objects, not numericals. In order to have more precies correlation alayses, first we need to transform those variables into numericals.

```python
le = LabelEncoder()
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
df['Embarked'] = le.fit_transform(df['Embarked'])
```

```python
def score_to_numeric(x):
    if x=='S':
        return 3
    if x=='C':
        return 2
    if x=='Q':
        return 1
    
df['Embarked'] = df['Embarked'].apply(score_to_numeric)
```

```python
df.drop(['PassengerId'], axis=1).corr().style.format('{:.2}').background_gradient(cmap=plt.get_cmap('PuBu'), axis=1)
```

```python
corrs_new = pd.DataFrame(df.corr()['Survived'][2:])
corrs_new.sort_values(by='Survived').style.format('{:.2}').background_gradient(cmap=plt.get_cmap('PuBu'))
```

As we expected, Ses variable is the highest correlated variable to survival, with negative value meaning that men had less chanse of survivial. We also transormed Embarked variable to numerical but it has not significant correlation to survival variable.
